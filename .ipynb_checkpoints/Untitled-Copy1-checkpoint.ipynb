```python
# Dependencies and Setup
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
#print(f"metadata: {mouse_metadata.shape}, study: {study_results.shape}")
mouse_metadata
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
study_results
```

```python
# Combine the data into a single dataset
merged_data_df = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")
merged_data_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
group_data_df = merged_data_df.groupby(["Drug Regimen"])
group_data_df.count()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
group_data_df = merged_data_df.groupby(["Drug Regimen"])
group_data_df.count().head()

mean = group_data_df["Tumor Volume (mm3)"].mean()

median = group_data_df["Tumor Volume (mm3)"].median()

var = group_data_df["Tumor Volume (mm3)"].var()

std_dev = group_data_df["Tumor Volume (mm3)"].std()

std_err = group_data_df["Tumor Volume (mm3)"].sem()
#mode = group_data_df["Tumor Volume (mm3)"].mode()
#mean_numpy = np.mean(merge_data)
#mean_numpy

#Calculating the summary of statistics
stats_df = pd.DataFrame({"Mean Tumor Volume":mean, "Median Tumor Volume":median, "Tumor Volume Variance":var, 
                         "Tumor Volume Std. Dev":std_dev, "Tumor Volume Std. Err":std_err})
stats_df
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas

count = group_data_df["Tumor Volume (mm3)"].count()

treat_reg_df = pd.DataFrame({"Data Points":count})
treat_sorted_df = treat_reg_df.sort_values("Data Points", ascending=False)

#Colors of the bar plot
#colors = ["blue", "pink", "red", "orange", "purple", "green", "grey", "light_blue", "yellow", "black"]
treat_sorted_df
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas
treat_sorted_df.plot(kind="bar", color="red",figsize=(6,4.5),legend=False, align="center")

# Aesthetics for the chart
plt.title("Treatment Regimen")
plt.ylabel("Number of Data Points")
plt.xlim(-.5,9.5)
plt.show()

#Saving an image of this plot
plt.savefig("Images/Pandas_DrugRegimen.png")
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pyplot
#Defining x and y
x_axis = treat_sorted_df.index
y_axis = treat_sorted_df["Data Points"]

#Plotting with pyplot
plt.bar(x_axis, y_axis, color='r', align="center", width=0.6)

# Aesthetics for the chart

plt.xticks(x_axis,rotation=90)
plt.title("Treatment Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Data Points")
plt.xlim(-.75,)
plt.show()

#Saving an image of this plot
plt.savefig("Images/Pyplot_DrugRegimen.png")
```

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens.
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

#Determine final tumor volume values 
merged_data_df
#Sort the merged data DF by Timepoint and group by Mouse ID
mouse_group = merged_data_df.sort_values("Timepoint", ascending=False).groupby(["Mouse ID"])

fin_tumor = mouse_group["Tumor Volume (mm3)"].first()
regimen = mouse_group["Drug Regimen"].first()

regimen_df = pd.DataFrame({"Tumor Volume (mm3)":fin_tumor, "Drug Regimen":regimen})
regimen_df
```

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens.
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

capomulin = regimen_df.loc[regimen_df["Drug Regimen"] == "Capomulin", ["Tumor Volume (mm3)"]]["Tumor Volume (mm3)"].quantile([0,0.25,0.5])
ramicane = regimen_df.loc[regimen_df["Drug Regimen"] == "Ramicane", ["Tumor Volume (mm3)"]]["Tumor Volume (mm3)"].quantile([0,0.25,0.5])
infubinol = regimen_df.loc[regimen_df["Drug Regimen"] == "Infubinol", ["Tumor Volume (mm3)"]]["Tumor Volume (mm3)"].quantile([0,0.25,0.5])
ceftamin = regimen_df.loc[regimen_df["Drug Regimen"] == "Ceftamin", ["Tumor Volume (mm3)"]]["Tumor Volume (mm3)"].quantile([0,0.25,0.5])

promising_reg_df = pd.DataFrame({"Capomulin Tumor Vol (mm3)":capomulin,"Ramicane Tumor Vol (mm3)":ramicane, 
                                 "Infubinol Tumor Vol (mm3)":infubinol, "Ceftamin Tumor Vol (mm3)":ceftamin})
promising_reg_df.index.name = "Quantiles"
promising_reg_df
# ramicane
# infubinol
# ceftamin
# capomulin
```

```python
#Create a combined data frame of all promising drug regimes
my_capomulin = regimen_df.loc[regimen_df["Drug Regimen"] == "Capomulin", ["Tumor Volume (mm3)"]]["Tumor Volume (mm3)"].tolist()
my_ramicane = regimen_df.loc[regimen_df["Drug Regimen"] == "Ramicane", ["Tumor Volume (mm3)"]]["Tumor Volume (mm3)"].tolist()
my_infubinol = regimen_df.loc[regimen_df["Drug Regimen"] == "Infubinol", ["Tumor Volume (mm3)"]]["Tumor Volume (mm3)"].tolist()
my_ceftamin = regimen_df.loc[regimen_df["Drug Regimen"] == "Ceftamin", ["Tumor Volume (mm3)"]]["Tumor Volume (mm3)"].tolist()

my_COMBINED = [my_capomulin, my_ramicane, my_infubinol, my_ceftamin]
# my_COMBINED_df = pd.DataFrame({"Tumor Vol":my_COMBINED})
# my_COMBINED_df

my_combined_df = pd.DataFrame({"Capomulin":my_capomulin, "Ceftamin":my_ceftamin, "Infubinol":my_infubinol, "Ramicane":my_ramicane})
my_combined_df.head(30)
```
