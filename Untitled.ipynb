```python
# Dependencies and Setup
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
mouse_metadata.head()
```

```python
study_results
```

```python
# Combine the data into a single dataset
merged_data_df = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])
merged_data_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
group_data_df = merged_data_df.groupby(["Drug Regimen"])
group_data_df.count()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
group_data_df = merged_data_df.groupby(["Drug Regimen"])
group_data_df.count().head()

mean = group_data_df["Tumor Volume (mm3)"].mean()

median = group_data_df["Tumor Volume (mm3)"].median()

var = group_data_df["Tumor Volume (mm3)"].var()

std_dev = group_data_df["Tumor Volume (mm3)"].std()

std_err = group_data_df["Tumor Volume (mm3)"].sem()
#mode = group_data_df["Tumor Volume (mm3)"].mode()
#mean_numpy = np.mean(merge_data)
#mean_numpy

#Calculating the summary of statistics
stats_df = pd.DataFrame({"Mean Tumor Volume":mean, "Median Tumor Volume":median, "Tumor Volume Variance":var, 
                         "Tumor Volume Std. Dev":std_dev, "Tumor Volume Std. Err":std_err})
stats_df
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas

count = group_data_df["Tumor Volume (mm3)"].count()

treat_reg_df = pd.DataFrame({"Data Points":count})
treat_sorted_df = treat_reg_df.sort_values("Data Points", ascending=False)
treat_sorted_df
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas
treat_sorted_df.plot(kind="bar", color="red",figsize=(7,5),legend=False, align="center")

# Aesthetics for the chart
plt.title("Treatment Regimen")
plt.ylabel("Number of Data Points")
plt.xlim(-.5,9.5)
plt.show()

#Saving an image of this plot
plt.savefig("../Images/Pandas_DrugRegimen.png")
```
